```python
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold
from sklearn import linear_model
from sklearn import datasets
```

why we need stratified k-fold cross validation over simple k-fold cross validation
in normal splitting of the dataset which consists of several classes of data (for now let's say class A and class B), if we split it, there are chances that the train split will have majority of class A and test split will have majority of class B.

this means that the model will only learn to classify Class A similar data more than that of those belonging to class B. 
ignoring this 80-20 rule, stratified k-fold cv helps in the following way :-
if there are
    - 80 samples of class 0
    - 20 samples of class 1

using random sampling in an 80:20 split then all 80 class 0 in the training set and all 20 class 1 in the test set.

using stratified sampling on the same dataset:

    1. Training set(80 samples):
        - 64 from class 0 (80% of 80)
        - 16 from class 1 (80% of 20)

    2. Test set(20 samples)
        - 16 from class 0 (20& of 80)
        - 4 from class 1 (20% of 20)
        

```python
cancer = datasets.load_breast_cancer()

x=cancer.data #feature/input values
y=cancer.target #ouput/class labels(0/1)
```

```python
#feature scaling (normalization)
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#model and K-fold object setup
lr = linear_model.LogisticRegression()

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
lst_accu_stratified = []
```

```python
#applying stratified K-Fold and Training model
for train_index, test_index in skf.split(x, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    lr.fit(x_train_fold, y_train_fold)
    lst_accu_stratified.append(lr.score(x_test_fold, y_test_fold))
    
```

```python
#printing the accuracy results
print("list of possible accuracy: ", lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
	max(lst_accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
	min(lst_accu_stratified)*100, '%')
print('\nOverall Accuracy:',
	mean(lst_accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```
